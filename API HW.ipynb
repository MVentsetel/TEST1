```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
KEYWORDS = ['python', 'парсинг']
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
posts = soup.find_all('article', class_ = 'tm-articles-list__item')
posts
```

```python
habr_post = pd.DataFrame()
posts = soup.find_all('article', class_ = 'tm-articles-list__item')
for post in posts:
    post_lower = post.text.lower()
    for word in KEYWORDS:
        if word.lower() in post_lower:
            date_keyword = pd.to_datetime(soup.find('time').get('datetime'), dayfirst=True).date()
            title_keyword = post.find('a', class_='tm-article-snippet__title-link')
            url_keyword = 'https://habr.com' + title_keyword.attrs.get('href')
            row = {'date': date_keyword, 'title': title_keyword, 'url': url_keyword}
            habr_post = pd.concat([habr_post, pd.DataFrame([row])])
habr_post
```
