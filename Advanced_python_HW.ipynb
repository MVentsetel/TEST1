```python
import pandas as pd
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()
check = ratings_count[ratings_count['rating'] >= 100]
best_users = check['userId'].tolist()
print(best_users)
```

```python
New_ratings = ratings.loc[ratings['userId'].isin(best_users)]
lifetime_user = New_ratings.groupby('userId').agg(['min', 'max'])['timestamp'].reset_index()
lifetime_user['min'] = pd.to_datetime(lifetime_user['min'], unit = 's')
lifetime_user['max'] = pd.to_datetime(lifetime_user['max'], unit = 's')
lifetime_user.head(200000)
```

```python
def minusator(row):
    x = row['max'] - row['min']
    return x

lifetime_user['timelife'] = lifetime_user.apply(minusator, axis=1)
lifetime_user.head(20)
```
