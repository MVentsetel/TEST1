```python
na_values = ['None', '(S)', '?']
horse_data = pd.read_csv('horse_data.csv', header=None, dtype={'pctapi': np.float64}, na_values=na_values)
horse_data.columns = true_names
filtered_horse_data = horse_data[['1:  surgery?', '2:  Age', '4:  rectal temperature', '5:  pulse', '6:  respiratory rate', '7:  temperature of extremities', "11: pain - a subjective judgement of the horse's pain level", '23: outcome']]
filtered_horse_data
```

```python
# большинство лошадей были прооперированы,
# большинство лошадей взрослые,
# средняя ректальная температура всех лошадей соответствует нормальной,
# пульс выше нормальных значений,
# средняя частота дыхания выше нормальной,
# средняя температура конечностей прохладная,
# средний уровень боли - прерывистая легкая боль,
# в итоге большинство лошадей выжило.
```

```python
q1 = filtered_horse_data['5:  pulse'].quantile(0.25)
q3 = filtered_horse_data['5:  pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = filtered_horse_data[filtered_horse_data['5:  pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('5:  pulse')
remove_outliers
# ищу выбросы по пульсу.
```

```python
pd.concat([filtered_horse_data, remove_outliers]).drop_duplicates(keep=False)
```

```python
new_filtered['7:  temperature of extremities'].fillna(new_filtered['7:  temperature of extremities'].mean(), inplace=True)
new_filtered['7:  temperature of extremities'].fillna(new_filtered['7:  temperature of extremities'].mean()).isna().sum()
```
