```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
merged_movie = movies.merge(ratings, how = 'inner', left_on='movieId', right_on='movieId')
merged_movie['rating_comma'] = merged_movie['rating'].apply(lambda x: str(x).replace('.', ','))
merged_movie.sort_values(by='rating_comma').head()
```

```python
def movie_class_rating(merged_movie):
    if merged_movie.rating_comma <= '2,0':
        return ('низкий рейтинг')
    elif merged_movie.rating_comma >= '4,5':
        return ('высокий рейтинг')
    else:
        return ('средний рейтинг')
```

```python
merged_movie['movie_class_rating'] = merged_movie.apply(movie_class_rating, axis=1)
merged_movie.head()
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def find_region(row):
    if 'москва' in row.keyword or 'тула' in row.keyword or 'ярославль' in row.keyword:
        return ('Center')
    if 'петербург' in row.keyword or 'псков' in row.keyword or 'мурманск' in row.keyword:
        return ('Northwest') 
    if 'владивосток' in row.keyword or 'сахалин' in row.keyword or 'хабаровск' in row.keyword:
        return ('FarEast')
    else:
        return ('undefined')
```

```python
keywords['region'] = keywords.apply(find_region, axis=1)
keywords.sort_values(by='region').head()
```

```python
df_Center = keywords[keywords.region == 'Center']
df_Center
```

```python
df_Northwest = keywords[keywords.region == 'Northwest']
df_Northwest
```

```python
df_FarEast = keywords[keywords.region == 'FarEast']
df_FarEast
```

```python
mov = pd.read_csv('movies.csv')
rat = pd.read_csv('ratings.csv')
merged = mov.merge(rat, how = 'inner', left_on='movieId', right_on='movieId')
merged['rating_comma'] = merged['rating'].apply(lambda x: str(x).replace('.', ','))
merged.sort_values(by='rating_comma').tail()
```

```python
import re
def true_year(merged):
    year = re.findall(r'\(\d\d\d\d', merged.title)
    if year != []:
        for i in year:
            return int(i.replace('(', ''))
    else:
        return 1900
```

```python
merged['true_year'] = merged.apply(true_year, axis=1)
merged.sort_values(by='true_year').head()
```

```python
mean_ratng_by_year = merged.groupby('true_year')
```

```python
mean_ratng_by_year.mean().sort_values(by='rating', ascending=0).head(30)
```
