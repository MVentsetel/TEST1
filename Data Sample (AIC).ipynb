```python
import pandas as pd
import numpy as np
online_retail = pd.read_csv('Online_Retail.csv')
online_retail.head()
# Читаю данные
```

```python
online_retail.info()
# смотрю тип данных и где их не хватает
```

```python
online_retail['UnitPrice'] = online_retail['UnitPrice'].apply(lambda x: str(x).replace(',', '.'))
online_retail['UnitPrice'] = online_retail['UnitPrice'].astype(float)
online_retail['Quantity'] = online_retail['Quantity'].astype(float)
online_retail.loc[:, 'earnings'] = online_retail['Quantity'] * online_retail['UnitPrice']
online_retail
# меняю тип данных для вычисления выручки
```

```python
new_data = online_retail.loc[online_retail['earnings'] > 0]
new_data
# без отрицательных величин и нулей
```

```python
new_data.groupby('Country').count().sort_values(by=['Description'], ascending = False)
# Смотрю разбивку по странам
```

```python
new_data.loc[new_data['Country'] == 'United Kingdom', 'Region'] = 'UK'
new_data.loc[new_data['Country'] != 'United Kingdom', 'Region'] = 'Export'
new_data
# Добавляю регион uk или экспорт
```

```python
new_data['InvoiceDate'] = pd.to_datetime(new_data['InvoiceDate'])
new_data
# делаю дату читаемой
```

```python
new_data.groupby('Region').count().sort_values(by=['Description'], ascending = False)
# Смотрю разбивку по регионам
```

```python
# new_data.loc[new_data['CustomerID'] == ' ', 'CustomerID'] == 'No data'
new_data['CustomerID'].fillna('No data', inplace=True)
new_data
```

```python
new_data.to_excel('Online_Retail_new.xlsx', index=False)
# записываю файл
```
